```python
data = pd.read_csv("../data/profiles.csv")
# data.head(1)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns

# Select only the numeric columns
numeric_data = data.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Display the correlation matrix using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

Chaque entre dans la matrice de corrlation reprsente le coefficient de corrlation entre deux variables. Le coefficient de corrlation est une mesure statistique qui quantifie le degr selon lequel deux variables voluent l'une par rapport  l'autre. Il varie de -1  1, o :

- 1 indique une corrlation positive parfaite (quand une variable augmente, l'autre variable augmente galement proportionnellement),
- -1 indique une corrlation ngative parfaite (quand une variable augmente, l'autre variable diminue proportionnellement),
- 0 indique aucune corrlation linaire.

# Interprtation:
- Aprs avoir examin la matrice de corrlation, on peut conclure qu'il existe une relation linaire entre la variable dpendante `Citations globales` et les variables indpendantes `H Indice global`, `Citations depuis 2018` et `H Indice depuis 2018`. Cela est indiqu par des coefficients de corrlation suprieurs  0.7, s'approchant de 1.
- De plus, il existe galement une forte corrlation entre les variables `Citations depuis 2018` et `H Indice depuis 2018` avec un coefficient de 0.93.
- Il existe une relation faible entre les variables `Citations depuis 2018` et `H indice_cl10 global` avec un coefficient de 0.52, ainsi qu'entre les variables `H Indice depuis 2018` et `H indice_cl10 global` avec un coefficient de 0.4.

# Interprtation :

- La direction et la forme des points indiquent la nature de la relation :
    - `Pente Positive` :  mesure qu'une variable augmente, l'autre a tendance  augmenter.
    - `Pente Ngative` :  mesure qu'une variable augmente, l'autre a tendance  diminuer.
    - `Aucun Modle Clair` : Il pourrait ne pas y avoir de relation linaire forte.

- La densit et la dispersion des points dans les graphiques de dispersion indiquent la force de la relation entre les variables.
- Si les points se regroupent troitement autour d'une ligne, cela suggre une corrlation forte.
  - par example `citaion global` et `H indec global`
- Lorsque les points dans un graphique de dispersion sont disperss sans tendance ou motif clair, cela suggre l'absence de relation linaire. Les variables peuvent tre faiblement corrles ou avoir une relation non linaire.
  - par example `H indie_Ccl10 global` avec `citaion global` et `H indec global`

```python

# nettoyage des donnes
numeric_data.fillna(numeric_data.mean(), inplace = True)
X1 = numeric_data[[
    'H indice depuis 2018',
    'Citations depuis 2018',
    'H indice global'
]]
y1 = numeric_data['Citations global']

X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size = 0.2, random_state = 42)

print(f"Ensemble d'entranement : ")
print(X_train)
print('_________________________________________________________________\n')
print(f"Ensemble de test : ")
print(y_train)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)

R1 = r2_score(y_test, y_pred)

print(f"Coefficient de dtermination -> R = {R1} ")
```

# Interprtation :
Le coefficient de dtermination (R) est trs proche de 1 (0,8586833570479674), ce qui signifie que votre modle explique pratiquement toute la variance de la variable dpendante (Ladder score).

En d'autres termes, environ 85,86% de la variabilit de la variable dpendante est explique par les variables indpendantes ('H indice depuis 2018', 'Citations depuis 2018', 'H indice global'). Il s'agit d'un ajustement extrmement prcis.
