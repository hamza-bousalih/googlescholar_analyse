```python
import pandas as pd
from scipy.stats import f_oneway
```

```python
# Create a DataFrame with your data
data = pd.read_csv("../data/profiles.csv")

df = pd.DataFrame(data)
df.head(5)
```

```python
# Define groups for the ANOVA test
groups = df["faculty"].unique()

# Check for missing values
print("Missing values:")
print(df.isnull().sum())

# Drop rows with missing values (you may choose a different approach)
df_clean = df.dropna()

# Perform ANOVA for H indice global
h_index_anova = f_oneway(*[df_clean[df_clean["faculty"] == group]["H indice global"] for group in groups])

# Perform ANOVA for Citations global
citations_anova = f_oneway(*[df_clean[df_clean["faculty"] == group]["Citations global"] for group in groups])

# Print ANOVA results
print("\nANOVA for H indice global:")
print(h_index_anova)

print("\nANOVA for Citations global:")
print(citations_anova)
```
