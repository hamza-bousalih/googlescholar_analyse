```python
import requests
from bs4 import BeautifulSoup
import csv
import json

    
headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/9.0.2 Safari/601.3.9'}

profiles=[]
links=[]

for i in 0, 10:
    url = f"https://scholar.google.com/citations?view_op=view_org&hl=fr&org=10696271937461036765&before_author=sslq_4kLAAAJ&astart={i}"
    headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/9.0.2 Safari/601.3.9'}
    page=requests.get(url,headers=headers)
    src=page.content
    soup=BeautifulSoup(src,"lxml")
    profiles.extend(soup.findAll('div', {'class': 'gsc_1usr'}))
for i in range(len(profiles)):
    links.append(profiles[i].find("a").attrs["href"])

# https://serpapi.com/search.json?engine=google_scholar&q=
NomProfil=" "
ArticleListe=[]
Statut_Depuis_2018=[]
Statut_global=[]
Assc_unevercity=[]
Specialty=""
ProfillDetaill=[]
 #  indice  Citations and indice H
def gettableofsat(soup):
    table_of_sat=soup.find(id= 'gsc_rsb_st')
    body_of_table=table_of_sat.contents[1]
    hs_indeces=body_of_table.contents[1].find_all("td",{'class' : 'gsc_rsb_std'})
    Citations=body_of_table.contents[0].find_all("td",{'class' : 'gsc_rsb_std'})
    h_indice_I10=body_of_table.contents[2].find_all("td",{'class' : 'gsc_rsb_std'})

    
    Statut_Depuis_2018.append({"H indice":hs_indeces[0].text.strip(),"Citations":Citations[0].text.strip(),"H indice_I10":hs_indeces[0].text.strip()})
    Statut_global.append({"H indice":hs_indeces[1].text.strip(),"Citations":Citations[1].text.strip(),"H indice_I10":h_indice_I10[1].text.strip()})
   

    
#le nom de profille
def getNameProfil(soup):
    NomProfil=soup.find(id = 'gsc_prf_in').text.strip()

# unevrcity
def getUnevercity(soup):
    assc_or_univer= soup.find("div",{'class' : 'gsc_prf_il'})
    Assc_unevercity= assc_or_univer.get_text(strip=True)

# spicialty
def spicialty(soup):
    spicialty=soup.find(id="gsc_prf_int")
    Specialty= spicialty.get_text(strip=True)

# all articles
def getArticle(soup):
    article= soup.find_all("tr",{'class' : 'gsc_a_tr'})

    print(len(article))
    for  i in range(len(article)):
        Areticletitle= article[i].find("a",{'class' : 'gsc_a_at'}).text.strip()
        AuteurAnddescription=article[i].find_all("div",{'class' : 'gs_gray'})
        auteurs=AuteurAnddescription[0].get_text(strip=True)
        description=AuteurAnddescription[1].get_text(strip=True)
        annee=article[i].find("td",{'class' : 'gsc_a_y'}).get_text(strip=True)
        ArticleListe.append({'Nom Article':Areticletitle, 'description':description,'auteur':auteurs,"Annee":annee})

def main(page):
    src=page.content
    soup=BeautifulSoup(src,"lxml")
    NomProfil=soup.find(id = 'gsc_prf_in').text.strip()
    spicialty=soup.find(id="gsc_prf_int")
    Specialty= spicialty.get_text(strip=True)
    assc_or_univer= soup.find("div",{'class' : 'gsc_prf_il'})
    Assc_unevercity= assc_or_univer.get_text(strip=True)
    getArticle(soup)
    spicialty(soup)
    getUnevercity(soup)
    getNameProfil(soup)
    gettableofsat(soup)
    
    info_0 = Statut_Depuis_2018[0]
    info_1 = Statut_global[0]
  
    ProfillDetaill.append({"Nom et Prénom":NomProfil,"ASS - Unevercity":Assc_unevercity,"Spciality":Specialty,
                           "H indice depuis 2018": info_0["H indice"],"Citations depuis 2018": info_0["Citations"],"H indice_I10 depuis 2018": info_0["H indice_I10"],"H indice global": info_1["H indice"],"Citations global": info_1["Citations"],"H indice_I10 global": info_1["H indice_I10"],"Article":ArticleListe})
    
    # print(ProfillDetaill)
   
    with open('uca.json', 'w') as outfile:
        json.dump(ProfillDetaill, outfile)

for i in range(len(profiles)):
    page = requests.get("https://scholar.google.com"+links[i], headers=headers)
    ArticleListe=[]
    Statut_Depuis_2018=[]
    Statut_global=[]
    Assc_unevercity=[]

    main(page)
```
